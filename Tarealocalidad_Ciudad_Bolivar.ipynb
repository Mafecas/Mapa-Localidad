```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Cargar el archivo GeoJSON
ruta_al_archivo_geojson = "barrios_prueba.geojson"
gdf = gpd.read_file(ruta_al_archivo_geojson)

# Mostrar las primeras filas del GeoDataFrame original para verificar
print("GeoDataFrame original:")
print(gdf.head())

# Mostrar los nombres de las columnas
print("Nombres de las columnas:")
print(gdf.columns)

# Especificar la localidad deseada
localidad_deseada = "Ciudad Bolívar"

# Filtrar el GeoDataFrame
gdf_filtrado = gdf[gdf['localidad'] == localidad_deseada]

# Mostrar las primeras filas del GeoDataFrame filtrado para verificar
print("\nGeoDataFrame filtrado:")
print(gdf_filtrado.head())

# Asegurarse de que el GeoDataFrame esté en el CRS correcto para contextily
gdf_filtrado = gdf_filtrado.to_crs(epsg=3857)

# Crear la figura y los ejes
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Plotear el GeoDataFrame filtrado
gdf_filtrado.plot(column='barriocomu', ax=ax, legend=True, legend_kwds={'bbox_to_anchor': (1, 1)})

# Añadir el mapa base utilizando contextily
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, zoom=12)

# Añadir la grilla con coordenadas
ax.grid(True, which='both', linestyle='--', color='gray', alpha=0.5)
ax.set_axisbelow(True)



# Añadir flecha del norte
x, y, arrow_length = 0.95, 0.1, 0.1
ax.annotate('N', xy=(x, y), xytext=(x, y - arrow_length),
            arrowprops=dict(facecolor='black', width=5, headwidth=15),
            ha='center', va='center', fontsize=15, xycoords=ax.transAxes)

# Título y ajustes finales
plt.title(f'Barrios de {localidad_deseada}')
plt.xlabel('Longitud')
plt.ylabel('Latitud')

# Mostrar el mapa
plt.show()
```
